```python
## Query 16- Stats leader for each stat
sql ='''


select 
'Highes Total:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by total desc
limit 1) stats


union all

select 
'Highes Attack:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by attack desc
limit 1) stats

union all

select 
'Highes Defense:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by defense desc
limit 1) stats

union all

select 
'Highes Special Attack:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by spattack desc
limit 1) stats

union all

select 
'Highes Special Defense:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by spdefense desc
limit 1) stats

union all

select 
'Highes Speed:-    '|| stats.species as Stats_Master
from (
select
species,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon
group by species
order by speed desc
limit 1) stats



''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query 18- Pokememon evolutions and stats

sql='''

select 
distinct
evolutions_table.Pokemon_Name,
evolutions_table.Pre_Evolved_State,
evolutions_table.Primary_Type,
hp, 
attack, 
defense, 
spattack, 
spdefense, 
speed,
total
from pokemon p
join (select 
distinct p1.species as Pokemon_Name,
p1.preevolution as Pre_Evolved_State,
p1.type1 as Primary_Type,
p1.type2 as Secondary_Type,
p1.ability1 as Primary_Ablility,
p1.abilityH as Secondary_Ablility
from pokemon p1
left join pokemon p2 on p1.species=p2.preevolution) evolutions_table
on p.species=evolutions_table.Pokemon_Name

''';


    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
## Query 19- Types of item

sql='''

select * from (
select
item,
case 
when item like '% Heal%' then 'Heal'
when item like '% potion%' then 'Potion'
when item like '% revive%' then 'Revive'
when item like '% berry%' then 'Berry'
when item like '%plate%' then 'Plate'
when item like '%tm%' then 'TM'
when item like '%stone%' then 'Stone'
when item like '%key%' then  'Key'
else 0 end as Classification
from items
) a
where a.classification !=0
''';


    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```
